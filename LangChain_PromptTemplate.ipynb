```python
# LangChain PromptTemplate
from langchain_core.prompts import PromptTemplate

explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

def explain_topic(topic):
    return explain_template.format(topic=topic)

# Test
print(explain_topic("AI"))
```

```python
# multi-input prompt system
multi_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

def generate_multi_prompt(topic, audience, tone):
    return multi_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Tests
print(generate_multi_prompt("AI", "beginners", "friendly"))
print(generate_multi_prompt("Python", "kids", "fun"))
print(generate_multi_prompt("Deep Learning", "engineers", "technical"))
```

```python
#Prompt Variations Engine
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

```python
#ChatPromptTemplate System
from langchain_core.prompts import ChatPromptTemplate
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role} who explains concepts clearly."),
    ("user", "Explain {topic}")
])

def chat_prompt(role, topic):
    return chat_template.format_messages(role=role, topic=topic)

# Test
print(chat_prompt("teacher", "Neural Networks"))
print(chat_prompt("interviewer", "Neural Networks"))
print(chat_prompt("motivator", "Neural Networks"))
```

```python
#Input Validation Layer
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience → setting default: beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone → setting default: casual")
        tone = "casual"

    return audience, tone
```

```python
# Prompt Generator App
style_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} step by step in a simple way"
    ),
    "interview": PromptTemplate(
        input_variables=["topic"],
        template="Ask 5 interview questions about {topic}"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} as a creative story"
    )
}

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    template = style_templates.get(style, style_templates["teaching"])
    base_prompt = template.format(topic=topic)

    return f"{base_prompt} | Audience: {audience}, Tone: {tone}"

# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
#Template Reusability Test
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

test_cases = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "casual"),
    ("Deep Learning", "expert", "formal"),
    ("Blockchain", "beginner", "fun"),
    ("Cybersecurity", "intermediate", "formal")
]

for topic, audience, tone in test_cases:
    print(reusable_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    ))
```

```python
def full_pipeline(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)
    prompt = generate_prompt(topic, audience, tone, style)
    return prompt

print(full_pipeline("AI", "beginner", "fun", "storytelling"))
```
